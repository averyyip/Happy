```python
chase = pd.read_csv("chase.csv", header=None)
cleaned_chase = chase.iloc[:,[0,2]]
cleaned_chase.iloc[:,0] = pd.to_datetime(cleaned_chase.iloc[:,0])
monthly_purchases = cleaned_chase.resample("M", on=0).sum()
data = [
    Scatter(
        x=monthly_purchases.index, # assign x as the dataframe column 'x'
        y=monthly_purchases.iloc[:,0] * -1
    )
]

layout = Layout(
    title='Cleaning',
    yaxis=dict(title='random distribution'),
    xaxis=dict(title='linspace')
)

fig = Figure(data=data, layout=layout)
iplot(fig)
```

```python
cleaned_purchases = purchases[["amount", "description", "purchase_date"]]
cleaned_purchases["purchase_date"] = pd.to_datetime(cleaned_purchases["purchase_date"])
monthly_purchases = cleaned_purchases.resample("M", on="purchase_date").sum()
data = [
    Scatter(
        x=monthly_purchases.index, # assign x as the dataframe column 'x'
        y=monthly_purchases["amount"]
    )
]

layout = Layout(
    title='Cleaning',
    yaxis=dict(title='random distribution'),
    xaxis=dict(title='linspace')
)

fig = Figure(data=data, layout=layout)
iplot(fig)
```
